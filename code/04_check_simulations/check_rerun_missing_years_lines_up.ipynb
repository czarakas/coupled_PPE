# Look at a single experiment

```python
# Compare new files to existing files
fpath_existing='/glade/campaign/cgd/tss/czarakas/'
fpath_new='/glade/scratch/czarakas/archive/'
PPE_code='COUP0020'
fname_existing1=PPE_code+'_PI_SOM_v02.clm2.h0.0149-02-01-00000.nc'
fname_existing2=PPE_code+'_PI_SOM_v02.clm2.h0.0159-02-01-00000.nc'
fname_new=PPE_code+'_PI_SOM_v03.clm2.h0.0154-02-01-00000.nc'
ds_existing1=xr.open_dataset(fpath_existing+PPE_code+'_PI_SOM_v02'+'/lnd/hist/'+fname_existing1)
ds_existing2=xr.open_dataset(fpath_existing+PPE_code+'_PI_SOM_v02'+'/lnd/hist/'+fname_existing2)
ds_new=xr.open_dataset(fpath_new+PPE_code+'_PI_SOM_v03'+'/lnd/hist/'+fname_new)
var='GPP'
```

```python
time2=ds_existing2['time.year']+ds_existing2['time.month']/12
time1=ds_existing1['time.year']+ds_existing1['time.month']/12
time_new=ds_new['time.year']+ds_new['time.month']/12
```

```python
diff=ds_existing2.GPP[0:36].values-ds_new.GPP[60:].values
print(np.nanmin(diff))
print(np.nanmax(diff))
```

```python
plt.figure(figsize=(10,5))
ilat=70
ilon=100
#ilat=45
#ilon=120
#ilat=85
#ilon=115
plt.plot(time_new,ds_new[var][:,ilat,ilon], 'o-',label='new')
plt.plot(time1, ds_existing1[var][:,ilat,ilon], '.-r',label='existing')
plt.plot(time2, ds_existing2[var][:,ilat,ilon], '.-r',label='existing')
plt.legend()
```

```python
# Write processed files
fdir_in='/glade/campaign/cgd/tss/czarakas/COUP0007_PI_SOM_v03/lnd/hist/'
fdir_out='/glade/campaign/cgd/tss/czarakas/COUP0007_PI_SOM_v03/lnd/proc/subset/'
fnames_to_process=['COUP0007_PI_SOM_v03.clm2.h0.0154-02-01-00000.nc',
                   'COUP0007_PI_SOM_v03.clm2.h1.0154-02-01-00000.nc', 
                   'COUP0007_PI_SOM_v03.clm2.h2.0154-02-01-00000.nc']
fnames_to_process_out=['COUP0007_PI_SOM_v02.clm2.h0.rerun_0154-02-01-00000.nc',
                       'COUP0007_PI_SOM_v02.clm2.h1.rerun_0154-02-01-00000.nc', 
                       'COUP0007_PI_SOM_v02.clm2.h2.rerun_0154-02-01-00000.nc']
```

```python
for i,fname in enumerate(fnames_to_process):
    fname_out=fnames_to_process_out[i]
    print(fname)
    print(fname_out)
    ds_in=xr.open_dataset(fdir_in+fname)
    ds_out=ds_in.isel(time=slice(0,60))
    ds_out.to_netcdf(fdir_out+fname_out, format='NETCDF3_64BIT')
```
