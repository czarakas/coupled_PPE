```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import glob
import pandas as pd
from scipy.stats import ttest_ind
from cartopy.util import add_cyclic_point
import cartopy.crs as ccrs
```

```python
import seaborn as sns
```

```python
ds=pd.read_csv('Statistical_significance_testing.csv')[0:36]
```

```python
plt.plot(ds.frac_changed_JjA.values,
        ds.frac_changed_DJF.values,'.')
```

```python
plt.subplot(2,2,1)
plt.plot(ds.frac_changed_annual.values,
         ds.frac_changed_JjA.values,'o',color='crimson',fillstyle='none')
plt.plot([0,0.5],[0,0.5],'-k')
plt.xlabel('Annual Fraction')
plt.ylabel('JJA Fraction')
plt.title('JJA')
plt.grid()

plt.subplot(2,2,3)
plt.plot(ds.frac_changed_annual.values,
         ds.frac_changed_SON.values,'o',fillstyle='none')
plt.plot([0,0.5],[0,0.5],'-k')
plt.xlabel('Annual Fraction')
plt.ylabel('SON Fraction')
plt.title('SON')
plt.grid()

plt.subplot(2,2,2)
plt.plot(ds.frac_changed_annual.values,
         ds.frac_changed_DJF.values,'o',color='darkorange',fillstyle='none')
plt.plot([0,0.5],[0,0.5],'-k')
plt.xlabel('Annual Fraction')
plt.ylabel('DJF Fraction')
plt.title('DJF')
plt.grid()

plt.subplot(2,2,4)
plt.plot(ds.frac_changed_annual.values,
         ds.frac_changed_MAM.values,'o',color='green',fillstyle='none')
plt.plot([0,0.5],[0,0.5],'-k')
plt.xlabel('Annual Fraction')
plt.ylabel('MAM Fraction')
plt.title('MAM')
plt.grid()

plt.tight_layout()
```

```python
sns.distplot(ds.frac_changed_annual.values, hist=False, label='Annual', color='gray',
            kde_kws=dict(linewidth=5))#, rug=True)

sns.distplot(ds.frac_changed_JjA.values, hist=False, label='JJA',color='crimson')#, rug=True)
sns.distplot(ds.frac_changed_SON.values, hist=False, label='SON',color='darkorange')#, rug=True)
sns.distplot(ds.frac_changed_DJF.values, hist=False, label='DJF',color='blue')#, rug=True)
sns.distplot(ds.frac_changed_MAM.values, hist=False, label='MAM', color='green')#, rug=True)

plt.xlabel('Fraction of Land with Significant Precip Changes')
plt.ylabel('Number of Ensemble Members')
plt.savefig('Seasonality_of_significance')
```
